### Finishing up for loops

```python
# First import the packages we need

import numpy as np

from matplotlib import pyplot as plt
from matplotlib import colors as colors
```

```python
#Make some arrays

a = np.arange(5)
b = np.sqrt(a)
c = a**3

# Make a for loop on the elements of the arrays, using zip

for (i,r) in zip(a,b):
    print('The square root of',i,'is',r)
    print(c[i])
    
```

```python
# Make an arrays of indixes using mgrid:

arr1,arr2 = np.mgrid[0:3, 0:3]

print(arr2.shape)

print(arr2)

# Do a for loop over arr2, a 2d array, using ndemumerate

for index, x in np.ndenumerate(arr2):
    print(index, x)

test = np.ndenumerate(arr2)

print(test)
```

## Plotting
.

```python
# Plotting examples
# colors in matplotlib: https://matplotlib.org/stable/users/dflt_style_changes.html

# Let's plot an array:

y1 = np.array([1, 2.4, 1.7, 0.3, 0.6, 1.8])

#Default line plot

plt.plot(y1)
plt.show()
```

```python
# Plot with symbols and color

plt.plot(y1,'r*')
plt.show()
```

```python
# Example with x and y data:

x1 = np.array([0.5,1,2,2.5,3,4])

plt.plot(x1,y1)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Example plot')
plt.show()
```

```python
# Slightly more interesting plots:


# Make a bunch of points:

x2 = np.linspace(0,10,101)   # 0 = start, 10=end, 101=number of points

#print(x2)

y2 = np.sin(x2)

plt.plot(x2,y2)
plt.show()

```

```python

# Two plots on same graph:

y3 = 2.*np.sin(x2/3.)

plt.plot(x2,y2)
plt.plot(x2,y3)
plt.show()

```

```python
# Make a log-log plot

y4 = x2**3
y5 = 10**(x2/4)

plt.plot(x2,y4,'.')
plt.plot(x2,y5)
plt.yscale('log')
plt.xscale('log')
plt.xlim([1.,10])
plt.ylim([1.,1000])
plt.show()
```
