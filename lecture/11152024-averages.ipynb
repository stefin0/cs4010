# Mean, standard deviation, and std of mean

<h3>

Last time we looked at some SPOT data.  Each question came with a mean and standard deviation, but is this really right?

Let's take a different example.  Below are the heights of men in my family:

[5'9", 5'10", 5'11", 5'11", 6', 6', 6'1", 6'3", 6'4"]

Let's calculate the mean and stddev, and make some plots

    
</h3>

```python
feet = np.array([5,5,5,5,6,6,6,6,6])
inches = np.array([9,10,11,11,0,0,1,3,4])

height_in = feet*12. + inches

print(height_in)

n = height_in.size

h_mean = np.mean(height_in)
h_sd = sd_2 = np.std(height_in,ddof=1)

print('mean = ',h_mean)
print('stddev = ',h_sd)

print('stddev of mean = ',h_sd/np.sqrt(n))

x = np.arange(68,78,0.1)
gauss_in = scipy_stats.norm.pdf(x,loc=h_mean,scale=h_sd)*n

plt.hist(height_in,bins=8)
plt.plot(x,gauss_in)
plt.show()
```

<h3>

This doesn't look too terrible.  But here we've measure each person's height to the nearest inch.  What if we could only measure to the nearest foot?

Rouding everyone's height to the nerest foot we get

[6',6',6',6',6',6',6',6',6']

Which has a mean of 6 feet (72 inches) and a standard deviation of zero.  

    
</h3>

```python

height_ft = np.round(height_in/12)

print(height_ft)

n = height_in.size

hf_mean = np.mean(height_ft)
hf_sd = sd_2 = np.std(height_ft,ddof=1)

print('mean = ',hf_mean)
print('stddev = ',hf_sd)

print('stddev of mean = ',hf_sd/np.sqrt(n))

x = np.arange(68,78,0.1)
gauss_ft = scipy_stats.norm.pdf(x,loc=hf_mean*12,scale=hf_sd*12)*n

plt.hist(height_ft,bins=30)
plt.plot(x/12,gauss_ft)
plt.plot(x/12.,gauss_in)

plt.show()
```

<h3>

This is clearly wrong.

We really need to take into account the precision of each measurement, not just the distribution.

When measuing to the nearest foot, each value is really $6' \pm 6''$.  So we can consider 6 inches the standard devation on each measurement?  Probaby not really, but let's use this as an example.

For our measurements to the nearest inch, the real values are something like $69 \pm 0.5$ inches, $70 \pm 0.5$ inches, etc.

Now for some math:

If we have some data set $x_i$ with $i=1...n$, the mean is:

$\langle x \rangle = \frac{1}{n} \sum{x_i}$

And the population standard deviation is:

$\sigma = \sqrt{\sigma^2} $

$\sigma^2 = \langle x - \langle x \rangle \rangle^2 $

$...$

$\sigma^2 = \langle x^2 \rangle - \langle x \rangle^2 $

$\sigma^2 = \frac{1}{n} \sum{x_i^2} - (\frac{1}{n} \sum{x_i})^2 $

And the sample standard deviation is:

$s = \sqrt{s^2} $

$s^2 = \frac{n}{n-1}\sigma^2 $


But now each value $x_i$ also has a standard deviation $\sigma_i$. This means each value is really:

$X_i \pm \Delta X_i = \langle x_i \rangle \pm \sigma_i $

From this, we can get:

$\langle x \rangle_i = X_i $

$\langle x^2 \rangle_i = \Delta X_i^2 - \langle x \rangle_i^2 = \sigma_i^2 - X_i^2 $

Then the mean is:

$\langle X \rangle = \frac{1}{n} \sum{X_i} $

And the population standard deviation is:

$\sigma_X = \sqrt{\sigma_X^2} $

$\sigma_X^2 = \langle X^2 \rangle - \langle X \rangle^2 $

$\sigma_X^2 = \frac{1}{n} \sum{(X_i^2 + \sigma_i^2)} - (\frac{1}{n} \sum{X_i})^2 $


</h3>

```python
# Let's make sure this works:

sigma2_in = 1/n*np.sum(height_in**2) - (1/n*np.sum(height_in))**2


std_i_in = np.zeros(n)+0.5

sigma2_corr_in = 1/n*np.sum(height_in**2+std_i_in**2) - (1/n*np.sum(height_in))**2

print('For measurements to nearest inch:')

print('sigma^2 = ',sigma2_in)
print('sigma^2 with measurment errors = ',sigma2_corr_in)

print('sample stddev = ',np.sqrt(sigma2_in*n/(n-1)),'inches')
print('sample stddev with measurement errors = ',np.sqrt(sigma2_corr_in*n/(n-1)),'inches')

# So for inches this standard deviation is just a little bigger.

print(' ')

sigma2_ft = 1/n*np.sum(height_ft**2) - (1/n*np.sum(height_ft))**2


std_i_ft = np.zeros(n)+0.5

sigma2_corr_ft = 1/n*np.sum(height_ft**2+std_i_ft**2) - (1/n*np.sum(height_ft))**2

print('For measurements to nearest foot:')

print('sigma^2 = ',sigma2_ft)
print('sigma^2 with measurment errors = ',sigma2_corr_ft)

print('sample stddev = ',np.sqrt(sigma2_ft*n/(n-1))*12,'inches')
print('sample stddev with measurement errors = ',np.sqrt(sigma2_corr_ft*n/(n-1))*12,'inches')

```

<h3>

This looks reasonable.  For measurement in inches, the stddev is just a little bigger.  For measurements in feet, it's about half a foot.

(It would be exactly half a foot if we used the population standard deviation.)

If all the errors on the measurements are the same, we could also find this by adding sigma from the data (without errors) and the sigma for each measurement in quadrature:

$\sigma_X^2 = \sigma_{data}^2 + \sigma_{measurement}^2 $

Or if they were different we could do it as a sum:

$\sigma_X^2 = \sigma_{data}^2 + \frac{1}{n} \sum{\sigma_i^2} $

    
</h3>

```python
# Example for height in inches

print('sigma^2 = ',sigma2_in)
print('sigma^2 with measurment errors = ',sigma2_corr_in)
print('sigma^2 with measurment errors = ',sigma2_in+0.5**2)
```

```python
# Test loading .csv files

# ~/Documents/csustan/RPT/SPOT_evaluations/Spring_2024-ASTR2100-001.csv
#data = np.loadtxt('/Users/brianmorsony/Documents/csustan/RPT/SPOT_evaluations/Spring_2024-ASTR2100-001.csv',skiprows=1,usecols=11,encoding='windows-1252')


#csv_files = ! ls /Users/brianmorsony/Documents/csustan/RPT/SPOT_surveys/*.csv
#csv_files = np.asarray(csv_files)
#print(csv_files)
#
#file = csv_files[0]


file = 'Spring_2024-ASTR2100-001.csv'

print(file)


#data = np.genfromtxt('/Users/brianmorsony/Documents/csustan/RPT/SPOT_evaluations/Spring_2024-ASTR2100-001.csv',skip_header=1,usecols=np.arange(11),delimiter=',',encoding='windows-1252',filling_values=np.nan)
#data = np.genfromtxt('/Users/brianmorsony/Documents/csustan/RPT/SPOT_evaluations/Spring_2024-ASTR3000-001.csv',skip_header=1,usecols=np.arange(11),delimiter=',',encoding='windows-1252',filling_values=np.nan)

data = np.genfromtxt(file,skip_header=1,usecols=np.arange(11),delimiter=',',encoding='windows-1252',filling_values=np.nan)

print(data.shape)

# The responses start at column 2

score1 = data[:,2]
score1 = score1[np.argwhere(score1>=1)].flatten()

# Use question 3 as our second data set.  Question 1 and 2 have the same mean.

score2 = data[:,4]
score2 = score2[np.argwhere(score2>=1)].flatten()

print(score1)
print(score2)

```

<h3>

So we loaded some data and have the responses to two questions.  

Let's see what the mean and standard devations are for each question:

</h3>

```python
mean_1 = np.mean(score1)
mean_2 = np.mean(score2)

sd_1 = np.std(score1,ddof=1)  # For a sample standard devation.
sd_2 = np.std(score2,ddof=1)

print('score1 mean =',mean_1)
print('score2 mean =',mean_2)

print('score1 std dev =',sd_1)
print('score2 std dev =',sd_2)

sd_of_mean_1 = sd_1/np.sqrt(score1.size)
sd_of_mean_2 = sd_2/np.sqrt(score2.size)

print('score1 std dev of the mean =',sd_of_mean_1)
print('score2 std dev of the mean =',sd_of_mean_2)

print(' ')

def sigma2(data,std):
    n = data.size
    std_i = np.zeros(n)+std
    sigma2 = 1/n*np.sum(data**2+std_i**2) - (1/n*np.sum(data))**2
    return(sigma2)

sigma2_1 = sigma2(score1,0.5)
sigma2_2 = sigma2(score2,0.5)

sd_corr_1 = np.sqrt(sigma2_1*n/(n-1))
sd_corr_2 = np.sqrt(sigma2_2*n/(n-1))

print('score1 std dev corr = ',sd_corr_1)
print('score2 std dev corr = ',sd_corr_2)

sd_of_mean_corr_1 = sd_corr_1/np.sqrt(score1.size)
sd_of_mean_corr_2 = sd_corr_2/np.sqrt(score2.size)

print('score1 std dev of the mean =',sd_of_mean_corr_1)
print('score2 std dev of the mean =',sd_of_mean_corr_2)

# Also works if we add error in quadrature, with a little correction:

print(' ')

n1=score1.size
print('stddev added in quadrature = ',np.sqrt( (sd_1**2 + 0.5**2) ) )
print('with corrections for sample stddev = ',np.sqrt( (sd_1**2*(n1-1)/n1 + 0.5**2) * n1/(n1-1) ) )
```

<h3>

So the stddev ends up a little bigger (about 25-30%) even for this data.

Note that applying this to the standard deviation of the mean really only makes sense if we KNOW our errors are Gaussian.

How about we look at out t-test data.  I don't know if there's a way to incorporate uncertainty in measurement directly, but what we can to is a simulation.  Rather than just use integer values, we can draw random values from our undertainty (say, assuming a uniform or normal distrubution) and add if ot each data point, the calculate our p-value for Welch's t-test.  We cna do this a bunch of times and look at the distrubution of p-values amd average p-value that we get.
    
</h3>


```python
csv_files = ! ls /Users/brianmorsony/Documents/csustan/RPT/SPOT_evaluations/*.csv
csv_files = np.asarray(csv_files)
print(csv_files)

file = csv_files[0]

file = 'Spring_2024-ASTR2100-001.csv'
#file = 'Spring_2024-PHYS4530-001.csv'

print(file)


#data = np.genfromtxt('/Users/brianmorsony/Documents/csustan/RPT/SPOT_evaluations/Spring_2024-ASTR2100-001.csv',skip_header=1,usecols=np.arange(11),delimiter=',',encoding='windows-1252',filling_values=np.nan)
#data = np.genfromtxt('/Users/brianmorsony/Documents/csustan/RPT/SPOT_evaluations/Spring_2024-ASTR3000-001.csv',skip_header=1,usecols=np.arange(11),delimiter=',',encoding='windows-1252',filling_values=np.nan)

data = np.genfromtxt(file,skip_header=1,usecols=np.arange(11),delimiter=',',encoding='windows-1252',filling_values=np.nan)

print(data.shape)

# The responses start at column 2

score1 = data[:,2]
score1 = score1[np.argwhere(score1>=1)].flatten()

# Use question 3 as our second data set.  Question 1 and 2 have the same mean.

score2 = data[:,4]
#score2 = data[:,5]
score2 = score2[np.argwhere(score2>=1)].flatten()

print(score1)
print(score2)



# Welch's t-test

mean_1 = np.mean(score1)
mean_2 = np.mean(score2)

print('score1 mean =',mean_1)
print('score2 mean =',mean_2)


diff_of_mean = mean_1-mean_2

print(' ')

welch_ttest = scipy_stats.ttest_ind(score1,score2,equal_var=False)

print(welch_ttest)

print(' ')

print('Difference of mean = ',diff_of_mean)

print('p-value = ',welch_ttest.pvalue)

print(' ')

print('Difference in mean',welch_ttest.confidence_interval(confidence_level=0.95))
```

```python
# Test out some simulation code

#print(score1)

n_1 = score1.size
n_2 = score2.size


repeat = 10000

mean_arr = np.zeros(repeat)
std_arr = np.zeros(repeat)
p_value_arr = np.zeros(repeat)

for i in np.arange(repeat):
    #sim_data_1 = score1 + np.random.uniform(-0.5,0.5,n_1)   # 0.87 to reproduce Guassian results
    #sim_data_2 = score2 + np.random.uniform(-0.5,0.5,n_1)
    sim_data_1 = score1 + np.random.normal(0.,0.5,n_1)
    sim_data_2 = score2 + np.random.normal(0.,0.5,n_2)
    mean_arr[i] = np.mean(sim_data_1) #-np.mean(sim_data_2)
    std_arr[i] = np.std(sim_data_1,ddof=1)
    welch_ttest = scipy_stats.ttest_ind(sim_data_1,sim_data_2,equal_var=False)
    p_value_arr[i] = welch_ttest.pvalue

print('average of mean of score1=',np.mean(mean_arr))
print('average stddev on score1 without correction =',np.mean(std_arr))
print('average stddev of mean of score1 without correction  =',np.std(mean_arr))



#print(np.mean(std_arr))

print('average p-value =',np.mean(p_value_arr))

plt.hist(mean_arr)
#plt.hist(std_arr)
plt.hist(p_value_arr,alpha=0.7)
plt.show()

    
```
