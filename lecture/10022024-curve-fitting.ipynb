## Curve fitting with non-polynomial functions

```python
# First, load any modules you need here.

import numpy as np
from matplotlib import pyplot as plt
from matplotlib import colors as colors

import scipy.optimize as optimize

```

<h3> 
    
General Curve Fitting


Now let look at some more interesting examples, using non-polynomial curves.  First, let's make some data and errors:


```python
# First, make some data:

n = 31  # number of points

x = np.linspace(0,n-1,n)  # make x from 0 to n-1 for now

np.random.seed(137)    # This seeds a random number generator.  Using a fixed seed makes out output reproducable.


# Make a sin wave with some point.  Period is 20., amplutide is 1., phase = 0

y = np.sin(x/(n-1)*np.pi*3) + np.random.randn(n)*0.3    # Add some random noise with sigma = 0.3

sigma = np.zeros(n)+0.3  # Set sigma to 0.3

plt.errorbar(x,y,sigma,marker='o',linestyle='')
plt.show()
```

```python
# Now let's try fitting a function to this data, using curve_fit:


def sin_fit(x,a,b,c):
    fit = a*np.sin(x*2*np.pi/b + c)
    return fit



p=[1.,20.,0.]  # initial guess

bounds = [[0,0,0],[np.inf,np.inf,2*np.pi]]

popt,pcov = optimize.curve_fit(sin_fit,x,y,p,sigma=sigma,bounds=bounds,absolute_sigma=True,method='dogbox')   # absolute_sigma=True is used if we know the errors on the data

perr = np.sqrt(np.diag(pcov))


print('popt = ',popt)
print('perr = ',perr)

#popt=[1.,20.,0]

fit = sin_fit(x,popt[0],popt[1],popt[2])

chisqr = np.sum((fit-y)**2/sigma**2)
red_chisqr = chisqr/(n-3)

print('chisqr = ',chisqr)
print('reduced chisqr = ',red_chisqr)

plt.errorbar(x,y,sigma,marker='o',linestyle='')
plt.plot(x,fit)
plt.show()
```

<h3> 
    
Poisson Error

Really, there will be an error on each point.  Since this is something like counts vs. frequency in some amount of time, the number of counts in each bin will follow a Poisson distribution.  In a given time, there is some expected value of count that we will get, $N$.  The standard deviation of this value will be $\sqrt{N}$.  

https://en.wikipedia.org/wiki/Poisson_distribution

In fact, I used the Poisson distrubution to make my number of counts per bin, above.  Let's assume Poisson errors on each point, and retry our fit:



```python

# Poisson errors should be the square root of the number of counts

sigma = np.sqrt(y)  # set error to new sqrt(y)

p=np.array([10.*norm,0.*norm,0.00*norm,10.*norm,3.,30.])  # initial guess


print(p)

#bounds = [[0,np.inf],[0,np.inf],[0,2*np.pi]]
#bounds = [[-np.inf,-np.inf,0,-np.inf,-np.inf,-np.inf],[np.inf,np.inf,1.e-5,np.inf,np.inf,np.inf]]
bounds = [[-np.inf,-np.inf,0,-np.inf,-np.inf,-np.inf],[np.inf,np.inf,1.e5,np.inf,np.inf,np.inf]]

#popt,pcov = optimize.curve_fit(sin_fit,x,y,p,sigma=sigma,bounds=bounds,absolute_sigma=True,method='dogbox')   # absolute_sigma=True is used if we know the errors on the data
popt,pcov = optimize.curve_fit(gauss_fit,x,y,p,sigma=sigma,absolute_sigma=True,bounds=bounds,method='dogbox')   # Use dogbox or trf # absolute_sigma=True is used if we know the errors on the data
#popt,pcov = optimize.curve_fit(gauss_fit,x,y,p,sigma=sigma,absolute_sigma=True,method='lm')   # use lm for no bounds  # absolute_sigma=True is used if we know the errors on the data

perr = np.sqrt(np.diag(pcov))


print('popt = ',popt)
print('perr = ',perr)


a = 10*norm
b = 0.01*norm
c = -0.001*norm
normal = 20.*norm
true_params = np.array([a,b,c,normal,3.,30])

print('true parameters = ',true_params)

print(' ')

print('line normalization = ',popt[3],'+/-',perr[3])
print('line center = ',popt[5],'+/-',perr[5])
print('line width = ',popt[4],'+/-',perr[4])

print(' ')

fit = gauss_fit(x,popt[0],popt[1],popt[2],popt[3],popt[4],popt[5])


chisqr = np.sum((fit-y)**2/sigma**2)
red_chisqr = chisqr/(n-6)

print('chisqr = ',chisqr)
print('reduced chisqr = ',red_chisqr)


plt.errorbar(x,y,sigma,marker='o',linestyle='')
plt.plot(x,fit)
#plt.plot(x,gauss_fit(x,a,b,c,normal,3.,30.))
plt.show()
```

<h3>

But this isn't really right either.  The number of counts in each bin includes the random error, so using the square root of that isn't the really error we should use.

Instead, we should use the sqaure root of the model fit at eahc point at the error.

There's no easy way I could figure out to do this with curve_fit, but I can to it with optimize.minimize.  However, getting the error bars is more compliacted and won't work for all solvers.

First, let's set up a function to use minimize to fit a curve to data assuming Poisson errors:


```python
# Set up the function we want to minimize:
    

def chi_squared_poisson(p,func,x,y):

    # func is the function to call to make a model.

    # return the weighted chi squared.
    
    fit = func(x,*p)   # use call the funciton with the given parameters.  The * is needed to pass the arguments correctly.

    
    sigma2 = np.abs(fit)  # sigma^2 is the absolute value of the value the function returns
    chisqr_w = np.sum((fit-y)**2/sigma2)   # calculate the weighted chi^2
    return chisqr_w



# Set up a new curve_fit function using minimize and poisson errors:


def curve_fit_poisson(func,x,y,p,bounds=None,method=None):

    # returns popt and pcov

    print(bounds)

    if (bounds == None):
        bounds2=bounds
    else:
        bounds2=[]
        for i in np.arange(len(bounds[0])):
            bounds2.append([bounds[0][i],bounds[1][i]])
            
    print(bounds2)
    
    best_fit = optimize.minimize(chi_squared_poisson,p,args=(func,x,y),bounds=bounds2,method=method)

    print(best_fit)
    
    popt = best_fit['x']

    fit = func(x,*popt)

    chisqr = np.sum((fit-y)**2/np.abs(fit))   # slightly different definition of chi squared.
    red_chisqr = chisqr/(x.size-len(popt))

    #print('test = ',np.sqrt(np.abs(best_fit.jac)))
    #print('test = ',np.abs(popt)*np.sqrt(np.abs(best_fit.jac)))
    #print('test = ',np.abs(best_fit.jac)*best_fit.fun/(len(y)-len(p)))
    #print('test = ',np.sqrt(np.abs(best_fit.jac)*best_fit.fun/(len(y)-len(p)))*popt)

    
    pcov = (best_fit.hess_inv * best_fit.fun/(len(y)-len(p)))

    '''
    J = best_fit.jac
    JT = J.T
    print(J,JT)
    print(JT.dot(J))
    pcov = np.linalg.inv(J.T.dot(J))
    '''
    
    return popt,pcov
```

```python
# Now use this function to do a new fit:

p=np.array([10.*norm,0.*norm,0.00*norm,10.*norm,3.,30.])  # initial guess


print(p)

#bounds = [[0,np.inf],[0,np.inf],[0,2*np.pi]]
#bounds = [[-np.inf,-np.inf,0,-np.inf,-np.inf,-np.inf],[np.inf,np.inf,1.e-5,np.inf,np.inf,np.inf]]
bounds = [[-np.inf,-np.inf,0,-np.inf,-np.inf,-np.inf],[np.inf,np.inf,1.e5,np.inf,np.inf,np.inf]]

#popt,pcov = optimize.curve_fit(sin_fit,x,y,p,sigma=sigma,bounds=bounds,absolute_sigma=True,method='dogbox')   # absolute_sigma=True is used if we know the errors on the data
#popt,pcov = optimize.curve_fit(gauss_fit,x,y,p,sigma=sigma,absolute_sigma=True,method='lm')   # use lm for no bounds  # absolute_sigma=True is used if we know the errors on the data

popt,pcov = curve_fit_poisson(gauss_fit,x,y,p)   # Use dogbox or trf # absolute_sigma=True is used if we know the errors on the data
#popt,pcov = curve_fit_poisson(gauss_fit,x,y,p,bounds=bounds,method='L-BFGS-B')   # Use dogbox or trf # absolute_sigma=True is used if we know the errors on the data

print(pcov)

perr = np.sqrt(np.diag(pcov))


print('popt = ',popt)
print('perr = ',perr)


a = 10*norm
b = 0.01*norm
c = -0.001*norm
normal = 20.*norm
true_params = np.array([a,b,c,normal,3.,30])

print('true parameters = ',true_params)

print(' ')

print('line normalization = ',popt[3],'+/-',perr[3])
print('line center = ',popt[5],'+/-',perr[5])
print('line width = ',popt[4],'+/-',perr[4])

print(' ')

fit = gauss_fit(x,popt[0],popt[1],popt[2],popt[3],popt[4],popt[5])


chisqr = np.sum((fit-y)**2/np.abs(fit))   # slightly different definition of chi squared.
red_chisqr = chisqr/(n-6)

print('chisqr = ',chisqr)
print('reduced chisqr = ',red_chisqr)


plt.errorbar(x,y,np.sqrt(np.abs(fit)),marker='o',linestyle='')
plt.plot(x,fit)
#plt.plot(x,gauss_fit(x,a,b,c,normal,3.,30.))
plt.show()
```

When we do the chi-squared fitting, we're still assuming symetrical Gaussian errors, but this isn't true for a Poisson distrubution.  It works okay as long at the number of counts is large, but breaks down for small values, and especially zero.
