# Gaussian Process Regression

<h3> Instead of using a spline or other interpolation, we are going to use a machine learning technique called Gaussian Process Regression.  (Sometimes also called Kriging.)
    
<br>
    
To do this, we will use the sklearn package, but you can do the same thing, with different syntax, using george or TensorFlow.
    
    

```python
# First, we need to load some data and get it into a format we can use.

# This is model data from a GRP afterglow code I've been working on:

file1 = './afterglow_sych_spectrum.GW170817.newfit.pass1.newwind.p'   # The datafile to load


# Set up some constant to use:

cl = 2.99792458e10  # speed fo light

h = 4.135667662e-15     #eV/s
to_ev = 4.1356691e-15 /2/np.pi  #convert omega(frequency) to eV
Mpc = 3.08e24  # Mpc in cm

z_origional = 0.047 # Redshift everything was run at

# For GW170817:
# Cosmological Redshift from Hjorth et al., 2017, ApJL, 848, L31
z = 0.0098
#For Ho = 73.24, OmegaM = 0.286, Omegavac = 0.714, z = 0.356, using Ned Wright's cosmology calculator.  
#This is the local Ho used by Hjorth et al. 2017, from Riess et al. 2016
dl = 40.4

# Frequencies to make some data at.  Just ignore this for now.

radio_f=np.array([6.5e8,1.e9,3.e9,6.e9,1.5e10,5.06e14,2.418e17,1.e22])*h/(to_ev*2*np.pi)
time_input='0'



# Load the data...

radio_fit1,xray_fit1,optical_fit1,optical_ab_fit1,time_fit1,freq_fit1, \
jy_fit1, time_origional_fit1, glow_fit1 \
= itpi.integrate_total_power_interp(file1, radio_f, z, z_origional, dl, \
                                    xray_min=0.3, xray_max=10.0, optical_f='0', time=time_input)


# The array we will want to use is jy_fit1, which is brightness in Janskies.

print('Shape of data array (jy_fit1) = ',jy_fit1.shape)

print('Shape of times (s) = ',time_fit1.shape)

print('Shape of frequencies (Hz) = ',freq_fit1.shape)
```

```python
# Plots some example data:


time = 48
freq = 70

print('time =',time_fit1[time]/1.e6,'* 10^6 s')
print('freq =',freq_fit1[freq]/1.e17,'* 10^17 Hz')

plt.plot(freq_fit1,jy_fit1[time,:])
plt.xscale('log')
plt.yscale('log')
plt.title('Spectrum at time = '+str(time_fit1[time])+'s')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Brightness (Jy)')
plt.show()



plt.plot(time_fit1,jy_fit1[:,freq])
plt.xscale('log')
plt.yscale('log')
plt.title('Lighcurve at freq = '+str(freq_fit1[freq])+'Hz')
plt.xlabel('Time (s)')
plt.ylabel('Brightness (Jy)')
plt.show()

```

```python
# Now let's try some interpolation.  

# Let's make a bunch more points in time and freqency and see how well interpolation works.

n = 1001 # number of points to interpolate at

new_time = np.logspace(np.log10(time_fit1[0]),np.log10(time_fit1[-1]),n)
new_freq = np.logspace(np.log10(freq_fit1[0]),np.log10(freq_fit1[-1]),n)


# We're going to fit at a reduced number of points:

ind_time = np.arange(0,time_fit1.size,2)
ind_freq = np.arange(0,freq_fit1.size,2)
print(freq_fit1[ind_freq].shape)

print(freq_fit1.size)
print(ind_freq)

# Now do 1-D interpolation to see how time and frequency look.


# Linear interpolation in 1-D:

interp_time_linear = scipy_interpol.interp1d(time_fit1[ind_time],jy_fit1[ind_time,freq],kind='linear')
interp_freq_linear = scipy_interpol.interp1d(freq_fit1[ind_freq],jy_fit1[time,ind_freq],kind='linear')


new_time_interp = interp_time_linear(new_time)
new_freq_interp = interp_freq_linear(new_freq)


# Interpolated at the missing points, so we can compare:

time_interp = interp_time_linear(time_fit1)
freq_interp = interp_freq_linear(freq_fit1)


plt.plot(freq_fit1,jy_fit1[time,:])
plt.plot(new_freq,new_freq_interp)
plt.xscale('log')
plt.yscale('log')
plt.title('At time = '+str(time_fit1[time]))
plt.xlabel('Frequency (Hz)')
plt.ylabel('Brightness (Jy)')
plt.show()

plt.plot(freq_fit1,jy_fit1[time,:]/freq_interp,'.')
plt.title('Difference between real and interpolated')
plt.xlabel('Frequency (Hz)')
plt.xscale('log')
plt.show()



plt.plot(time_fit1,jy_fit1[:,freq])
plt.plot(new_time,new_time_interp)
plt.xscale('log')
plt.yscale('log')
plt.title('At freq = '+str(freq_fit1[freq]))
plt.xlabel('Time (s)')
plt.ylabel('Brightness (Jy)')
plt.show()

plt.plot(time_fit1,jy_fit1[:,freq]/time_interp,'.')
plt.title('Difference between real and interpolated')
plt.xlabel('Time (s)')
plt.xscale('log')
plt.show()
```

# <br><br><br> Gaussian Process Regression


<h3> Let's use a machine learning technique instead.  This will not just do an interpolation, but will also give use error estimates on the points between data points.
    
<br>
    
It can also be used for data with known errors, taking those into account when interpolating.
 
<br>
    
Details on GPR can be found here:  https://scikit-learn.org/stable/modules/gaussian_process.html
    
    


```python
# Set up the GPR:

# We are going to do this on the whole 2D data set, not just in 1D.


# Get the 'X' values - time and frequency

Xt,Xf = np.meshgrid(np.log(time_fit1[ind_time]),np.log(freq_fit1[ind_freq]))
X = np.array([Xt.flatten(),Xf.flatten()]).T


# Scale them over a good range.

scaler = preprocessing.MinMaxScaler()
X_scaled = scaler.fit_transform(X)


# Get the 'Y' values - jy_fit1

ind_t,ind_f = np.meshgrid(ind_time,ind_freq)

y = np.log(jy_fit1[ind_t,ind_f]) #     y = np.log(jy.T) # Don't need the .T ??
y = y.flatten()


# Choose a kernel

kernel = Matern(nu=1.5)  # Okay overall.
#kernel = RationalQuadratic()  # SLOW
#kernel = ExpSineSquared()  # Good in the middle.
#kernel = RBF(length_scale=1.0, length_scale_bounds=(1e-1, 10.0))  # Doesn't work so well.
#kernel = DotProduct() # Fails.

#kernel = WhiteKernel()
#kernel = Matern(nu=1.5)+ConstantKernel()
#kernel = Matern(nu=1.5)*ExpSineSquared()

# Create the gpr procedure to be called for interpolation:

gpr_2D = GaussianProcessRegressor(kernel=kernel,random_state=0,n_restarts_optimizer=9) #,alpha=0.01**2.)
gpr_2D.fit(X_scaled,y)
gpr_2D.score(X_scaled,y)

print('Done making GPR')
```

```python
#=================================

# Now interpolate for new_time

# Create an array of frequencies at these time, all the same:
freq_arr = np.zeros(new_time.size)
freq_arr[:] = freq_fit1[freq]

# Make 'X' data to interpolate at:

XX_data = np.array([np.log(new_time),np.log(freq_arr)]).T
XX_data_scaled = scaler.transform(XX_data)

# Do the interpolation:

flux_log, sigma_new_time = gpr_2D.predict(XX_data_scaled, return_std=True)
new_time_interp_gpr = np.exp(flux_log)




# And for time_fit1:

freq_arr = np.zeros(time_fit1.size)
freq_arr[:] = freq_fit1[freq]

XX_data = np.array([np.log(time_fit1),np.log(freq_arr)]).T
    
XX_data_scaled = scaler.transform(XX_data)
    
flux_log, sigma_time = gpr_2D.predict(XX_data_scaled, return_std=True)
time_interp_gpr = np.exp(flux_log)




# And do the same for frequency:

time_arr = np.zeros(new_freq.size)
time_arr[:] = time_fit1[time]

XX_data = np.array([np.log(time_arr),np.log(new_freq)]).T
XX_data_scaled = scaler.transform(XX_data)

flux_log, sigma_new_freq = gpr_2D.predict(XX_data_scaled, return_std=True)
new_freq_interp_gpr = np.exp(flux_log)

# for time_fit1:

time_arr = np.zeros(freq_fit1.size)
time_arr[:] = time_fit1[time]

XX_data = np.array([np.log(time_arr),np.log(freq_fit1)]).T
    
XX_data_scaled = scaler.transform(XX_data)
    
flux_log, sigma_freq = gpr_2D.predict(XX_data_scaled, return_std=True)
freq_interp_gpr = np.exp(flux_log)





# Now plots:

plt.plot(freq_fit1,jy_fit1[time,:])
plt.plot(new_freq,new_freq_interp_gpr)
plt.xscale('log')
plt.yscale('log')
plt.title('At time = '+str(time_fit1[time]))
plt.xlabel('Frequency (Hz)')
plt.ylabel('Brightness (Jy)')
plt.show()

plt.plot(freq_fit1,jy_fit1[time,:]/freq_interp_gpr,'.')
plt.title('Difference between real and interpolated')
plt.xlabel('Frequency (Hz)')
plt.xscale('log')
plt.show()


plt.plot(time_fit1,jy_fit1[:,freq])
plt.plot(new_time,new_time_interp_gpr)
plt.xscale('log')
plt.yscale('log')
plt.title('At freq = '+str(freq_fit1[freq]))
plt.xlabel('Time (s)')
plt.ylabel('Brightness (Jy)')
plt.show()

plt.plot(time_fit1,jy_fit1[:,freq]/time_interp_gpr,'.')
plt.title('Difference between real and interpolated')
plt.xlabel('Time (s)')
plt.xscale('log')
plt.show()
```

<h3> Actually, this doesn't look too different from doing a cubic spline for this data.  But we can also take advantage of the sigma (error) that the GPR interpolation returns:

```python
# Use the plt.fill_between function to shade the 1-sigma region


plt.plot(freq_fit1,jy_fit1[time,:])
plt.plot(new_freq,new_freq_interp_gpr)

plt.fill_between(new_freq,np.exp(np.log(new_freq_interp_gpr)-sigma_new_freq),np.exp(np.log(new_freq_interp_gpr)+sigma_new_freq),alpha=0.5)   

plt.xscale('log')
#plt.yscale('log')
plt.title('At time = '+str(time_fit1[time]))
plt.xlabel('Frequency (Hz)')
plt.ylabel('Brightness (Jy)')
plt.show()


plt.plot(time_fit1,jy_fit1[:,freq])
plt.plot(new_time,new_time_interp_gpr)
plt.fill_between(new_time,np.exp(np.log(new_time_interp_gpr)-sigma_new_time),np.exp(np.log(new_time_interp_gpr)+sigma_new_time),alpha=0.5)   
plt.xscale('log')
#plt.yscale('log')
plt.title('At freq = '+str(freq_fit1[freq]))
plt.xlabel('Time (s)')
plt.ylabel('Brightness (Jy)')
plt.show()
```

```python
plt.plot(freq_fit1,jy_fit1[time,:]/freq_interp_gpr)
plt.fill_between(freq_fit1,jy_fit1[time,:]/np.exp(np.log(freq_interp_gpr)-sigma_freq),jy_fit1[time,:]/np.exp(np.log(freq_interp_gpr)+sigma_freq),alpha=0.5)   
plt.title('Difference between real and interpolated')
plt.xlabel('Frequency (Hz)')
plt.xscale('log')
plt.show()


#plt.plot(new_freq,np.exp(sigma_new_freq))
plt.plot(freq_fit1,np.exp(sigma_freq),'o')
plt.xscale('log')
plt.title('sigma_freq')
plt.xlabel('Frequency (Hz)')
plt.show()


plt.plot(time_fit1,jy_fit1[:,freq]/time_interp_gpr)
plt.fill_between(time_fit1,jy_fit1[:,freq]/np.exp(np.log(time_interp_gpr)-sigma_time),jy_fit1[:,freq]/np.exp(np.log(time_interp_gpr)+sigma_time),alpha=0.5)   
plt.title('Difference between real and interpolated')
plt.xlabel('Time (s)')
plt.xscale('log')
plt.show()


plt.plot(time_fit1,np.exp(sigma_time),'o')
plt.title('sigma_time')
plt.xlabel('Time (s)')
plt.xscale('log')
plt.show()
```

# <br><br><br><br>

<h3> GPR can also be used if you have errors on your input data.  For example this data isn't 100% accurate.  Let's say each point has a sigma of 1%.

```python
# Add some noise to the data:

sigma_data = np.copy(jy_fit1)
sigma_data[:] = 1.e-2

# Make a new GPR:

# Get the 'X' values - time and frequency

Xt,Xf = np.meshgrid(np.log(time_fit1[ind_time]),np.log(freq_fit1[ind_freq]))
X = np.array([Xt.flatten(),Xf.flatten()]).T


# Scale them over a good range.

scaler = preprocessing.MinMaxScaler()
X_scaled = scaler.fit_transform(X)


# Get the 'Y' values - jy_fit1

ind_t,ind_f = np.meshgrid(ind_time,ind_freq)

y = np.log(jy_fit1[ind_t,ind_f]) #     y = np.log(jy.T) # Don't need the .T ??
y = y.flatten()

alpha = np.log(1+sigma_data[ind_t,ind_f]**2) #  Calculate alpha = sigma^2 in log space
alpha = alpha.flatten()


# Choose a kernel

kernel = Matern(nu=1.5) 

# Create the gpr procedure to be called for interpolation:

gpr_2D = GaussianProcessRegressor(kernel=kernel,random_state=0,n_restarts_optimizer=9,alpha=alpha) #,alpha=0.01**2.)
gpr_2D.fit(X_scaled,y)
gpr_2D.score(X_scaled,y)

print('Done making GPR')
```
