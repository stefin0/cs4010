```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import colors as colors
```

```python
# Make some arrays
a = np.arange(5)
print(f"a = {a}")

b = np.sqrt(a)
print(f"b = {b}")

c = a**3
print(f"c = {c}")

# Make a for loop on the elements of the arrays, using zip
for (i, r) in zip(a, b):
    print(f"The square root of {i} is {r}")
    print(f"c[i] = {c[i]}")
```

```python
# Make an arrays of incies using mgrid:
arr1, arr2 = np.mgrid[0:3, 0:3]

# mgrid gives 2 arrays, with the x and y index (for 2 dimensions)
print(f"arr2.shape = {arr2.shape}")
print(f"arr2 =\n{arr2}\n")

# Do a for loop over arr2, a 2d array, using ndenumerate
for index, x in np.ndenumerate(arr2):
    print(f"index = {index}\nx = {x}\n")

test = np.ndenumerate(arr2)
print(f"test = {test}")
```

```python
# Plotting examples

# Let's plot an array:
y1 = np.array([1, 2.4, 1.7, 0.3, 0.6, 1.8])

# Default line plot
plt.plot(y1)
plt.show()
```
