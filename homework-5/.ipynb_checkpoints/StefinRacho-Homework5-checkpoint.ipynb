```python
# First, load any modules you need here.
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import colors as colors
import scipy.optimize as optimize
```

## 1) Line and Polynomial Fitting

The code below will generate some locatons, x1, data points, data1, and error on the data, sigma1.

```python
n = 20

x1 = np.linspace(0,n-1,n)  # make x from 0 to n-1 for now

np.random.seed(42)    # This seeds a random number generator.  Using a fixed seed makes out output reproducable.

data1 = 4. + 2.*x1 + 0.05*x1**2. + np.random.randn(n)*(x1*.3+1)

sigma1 = (x1*.1+1)

plt.errorbar(x1,data1,sigma1,marker='o',linestyle='')
plt.show()
```

Using this data, answer the following questions:

a) If you ignore the error and do a linear fit, what value of m and b do you get?  What are the errors on each of these values?

```python
def f(x,m,b):
    return m*x+b

fit_params, covariance = np.polyfit(x1, data1, 1, cov=True)

m, b = fit_params

errors = np.sqrt(np.diag(covariance))
m_error, b_error = errors

fit = f(x1, m, b)
plt.plot(x1, data1, "o")
plt.plot(x1, fit)
plt.show()
print('best fit m and b = ',m,b)
print(f"Error for m: {m_error}")
print(f"Error for b: {b_error}")
```

b) If you include the errors, what values of m and b do you get?  What are the errors on these values?  How close are these values to the ones used to create the data (m = 2, b = 4)?  Are the fitted values within 1-sigma of the real values?

```python
sigma = data1*0.1
weight = 1/sigma

fit_params, covariance = np.polyfit(x1, data1, 1, w=1/sigma1, cov=True)

m, b = fit_params

errors = np.sqrt(np.diag(covariance))
m_error, b_error = errors

fit = f(x1, m, b)
plt.plot(x1, data1, "o")
plt.errorbar(x1,data1,sigma1,marker='o',linestyle='')
plt.plot(x1, fit)
plt.show()

real_m, real_b = 2, 4
m_diff = abs(m - real_m)
b_diff = abs(b - real_b)

m_within_1sigma = m_diff <= m_error
b_within_1sigma = b_diff <= b_error

print('Best fit m and b (with errors included) =', m, b)
print(f"Error for m: {m_error}")
print(f"Error for b: {b_error}")
print(f"Is m within 1-sigma of real value (m = 2)? {m_within_1sigma}")
print(f"Is b within 1-sigma of real value (b = 4)? {b_within_1sigma}")
```
